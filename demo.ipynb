# LightGlue-ONNX Demo
This demo shows how to export LightGlue to ONNX and perform inference with ONNXRuntime.

```python
from pathlib import Path

from lightglue_dynamo.cli import export, infer
from lightglue_dynamo.config import Extractor, InferenceDevice
```

## ONNX Export
In this example we use SuperPoint features combined with LightGlue.

```python
extractor_type = Extractor.superpoint  # Extractor.disk
model_path = Path(f"weights/{extractor_type}_lightglue_pipeline.onnx")

export(extractor_type=extractor_type, output=model_path, batch_size=0, height=0, width=0, num_keypoints=1024)
```

## ONNX Inference

```python
infer(
    model_path=model_path,
    left_image_path=Path("assets/DSC_0410.JPG"),
    right_image_path=Path("assets/DSC_0411.JPG"),
    extractor_type=extractor_type,
    height=512,
    width=512,
    device=InferenceDevice.cuda,
)
```
